I got the data from this site: http://lib.stat.cmu.edu/datasets/IQ_Brain_Size

```python
import csv
from pylab import *

f = open("brainiq.csv","rb")
r = csv.reader(f)

a = []  # temporary list
iq = [] # full scale iq
bv = [] # total brain volume


for row in r:
    a = row[0].split()
    a = a[1:]
    a=map(float,a)
    iq.append(a[1])
    bv.append(a[7])
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Total Brain Volume over Full Scale IQ')
ax.set_xlabel('Total Brain Volume')
ax.set_ylabel('Full Scale IQ')
scatter(iq,bv)
savefig("fsiq.png")
    

    
f.close()
```

My conclusion is that there are no relations between the IQ scale and the brain volume
